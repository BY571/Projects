```python
import numpy as np 
import glob
import pandas as pd
```

```python
from glob import glob

images = []
images_names =[]
for fn in glob("Deckel/Kanaldeckel*.png"):
    print(fn)
    images_names.append(fn)
```

```python
print(len(images_names))
```

```python
df = pd.DataFrame(images_names)
```

```python
df["type"] = 1
```

```python
df.columns = ["image", "type"]
```

```python
r_images_names =[]
for i in glob("Random/*.png"):
    print(i)
    r_images_names.append(i)
```

```python
df2 = pd.DataFrame(r_images_names)
df2["type"]=0
df2.columns = ["image","type"]
```

```python
frames = [df,df2]
dataset = pd.concat(frames)
```
