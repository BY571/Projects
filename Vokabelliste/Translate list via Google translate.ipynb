```python
file = open('list.pkl', "rb")
old_list = pickle.load(file)
```

```python
#DataFrame erstellen:
data = pd.DataFrame(columns = ["Englisch","Deutsch"])
```

```python
# Liste in DataFrame einlesen
data["Englisch"] = old_list
```

```python
# Pro Element im Dataframe übersetzung suchen

translator = Translator()

for i in range(len(data["Englisch"])):
    #print(data["Englisch"][i])
    übersetzung = translator.translate(data["Englisch"][i],dest = "de")
    data["Deutsch"][i] = übersetzung.text
```
