# Preprocessing the data

```python
data.read()
```

```python
data_string = ""
for i in data:
    print(i)
    data_string = data_string+i
print("####################--CLEANING--#############################")
cleaned_data = data_string.replace("\n"," ")
print(cleaned_data)
```

```python
word_list = cleaned_data.split(" ")
```

```python
word_list
```

```python
from translate_api.translate_api import api
import unicodedata
src = "es"
dest = "de"
destination_word_list = []

for i in range(len(word_list)):
    try:
        translator = Translator()
        word_translation = translator.translate(word_list[i], src= src, dest= dest).text
    except:
        word_translation = api(text=word_list[i],from_language=src,to_language=dest)#,host=’https://translate.google.cn’, proxy=None
    destination_word_list.append(word_translation)
    
    
```

```python
destination_word_list
```

```python
import pandas as pd
```

```python
data_dic = {"Original_data": word_list, "Translated_data": destination_word_list}
data_frame = pd.DataFrame(data_dic)
```
